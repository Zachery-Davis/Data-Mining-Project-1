```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
aiMarket_df = pd.read_csv("ai_job_market_insights.csv")
aiMarket_df.head()
```

```python
aiMarket_df.isna().sum()
```

```python
aiMarket_df.dtypes
```

```python
aiMarket_df['Job_Title'] = aiMarket_df['Job_Title'].astype('string')
aiMarket_df['Industry'] = aiMarket_df['Industry'].astype('string')
aiMarket_df['Location'] = aiMarket_df['Location'].astype('string')
aiMarket_df['Required_Skills'] = aiMarket_df['Required_Skills'].astype('string')
aiMarket_df.dtypes
```

```python
mapping = {"Yes" : True, "No" : False}
aiMarket_df["Remote_Friendly"] = [mapping[value] for value in aiMarket_df["Remote_Friendly"]]
mapping = {"Growth" : 1, "Stable" : 0, "Decline" : -1}
aiMarket_df["Job_Growth_Projection"] = [mapping[value] for value in aiMarket_df["Job_Growth_Projection"]]
mapping = {"Large" : 3, "Medium" : 2, "Small" : 1}
aiMarket_df["Company_Size"] = [mapping[value] for value in aiMarket_df["Company_Size"]]
mapping = {"High" : 3, "Medium" : 2, "Low" : 1}
aiMarket_df["AI_Adoption_Level"] = [mapping[value] for value in aiMarket_df["AI_Adoption_Level"]]
aiMarket_df["Automation_Risk"] = [mapping[value] for value in aiMarket_df["Automation_Risk"]]
aiMarket_df.dtypes
```

```python
sns.set(rc={"figure.figsize":(30,12)})
sns.barplot(data=aiMarket_df, x='AI_Adoption_Level', y='Job_Title')
```

```python
sns.set(rc={"figure.figsize":(5,5)})
plt.grid(color='grey')
sns.boxplot(data=aiMarket_df, x='AI_Adoption_Level', y='Salary_USD', palette='muted')
```

```python
sns.set(rc={"figure.figsize":(10,5)})
sns.barplot(data=aiMarket_df, x='AI_Adoption_Level', y='Required_Skills')
```

```python
sns.set(rc={"figure.figsize":(8,6)})
plt.grid(color='grey')
sns.countplot(x='AI_Adoption_Level', hue='Automation_Risk', data=aiMarket_df, palette='muted')
plt.title('AI Adoption Level vs. Automation Risk')
plt.xlabel('AI Adoption Level')
plt.ylabel('Count')
plt.legend(title='Automation Risk')
plt.show()
```

```python
sns.set(rc={"figure.figsize":(10,6)})
plt.grid(color='grey')
sns.boxplot(x='AI_Adoption_Level', y='Salary_USD', hue='Automation_Risk', data=aiMarket_df, palette='muted')
plt.title('Salary Distribution by AI Adoption Level and Automation Risk')
plt.xlabel('AI Adoption Level')
plt.ylabel('Salary (USD)')
plt.legend(title='Automation Risk')
plt.show()
```

```python
sns.set(rc={"figure.figsize":(8,6)})
plt.grid(color='grey')
sns.countplot(x='AI_Adoption_Level', hue='Job_Growth_Projection', data=aiMarket_df, palette='muted')
plt.title('AI Adoption Level and Job Growth Projection')
plt.xlabel('AI Adoption Level')
plt.ylabel('Count')
plt.legend(title='Job Growth Projection')
plt.show()
```
